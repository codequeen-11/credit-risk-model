```python
from xverse.transformer import WOE
import pandas as pd

import os
os.getcwd()
import sys
sys.path.append("c:/credit-risk-model")

df = pd.read_csv("../data/raw/data.csv")
```

```python
from src.default_estimator import create_proxy_target

rfm = create_proxy_target(df)

rfm.head()
```

```python
categorical_cols = [
    "CurrencyCode",
    "ProviderId",
    "ProductCategory",
    "ChannelId",
    "PricingStrategy"
]
```

```python
target_df = rfm[
    ["CustomerId", "is_high_risk"]
]
```

```python
df_target = df.merge(
    target_df,
    on="CustomerId",
    how="left"
)
```
